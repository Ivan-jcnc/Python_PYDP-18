### Вычислить вектор градиента и матрицу Гессе для функции в точке M=(2,1,1)

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x1 = sym.Symbol('x1')
x2 = sym.Symbol('x2')
x3 = sym.Symbol('x3')
```

```python
Fx = 2*x1**3 - 3*x1*x2 + 2*x2*x3**2 - x2**3
Fx
```

Градиент - вектор частных производных

```python
gradFx = sym.Matrix([[sym.diff(Fx, x1,1)],
                  [sym.diff(Fx, x2,1)],
                  [sym.diff(Fx, x3,1)]
])
gradFx
```

Вычисление вектора градиента в точке M(2,1,1)

```python
gradFx.subs({x1:2,x2:1,x3:1})
```

Матрица Гессе

```python
Gesse = sym.Matrix([
    [sym.diff(Fx, x1,2), sym.diff(Fx, x1,x2), sym.diff(Fx, x1,x3)],
    [sym.diff(Fx, x2,x1), sym.diff(Fx, x2,2), sym.diff(Fx, x2,x3)],
    [sym.diff(Fx, x3,x1), sym.diff(Fx, x3,x2), sym.diff(Fx, x3,2)]
])

Gesse
```

Матрица Гессе в точке M(2,1,1)

```python
Gesse.subs({x1:2,x2:1,x3:1})
```

### Постройте график трехмерный график функции. А также график линий уровня с уровнями {1,5,10}

```python
F2x = x2**2 - x1**2*x2 + 2*x1*x2
F2x
```

```python
def func(x1,x2):
    return x2**2 - (x1**2)*x2 + 2*x1*x2
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = func(X,Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-100, 100)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.azim = 30
ax.dist = 10
ax.elev = 12

plt.show()
```

График линий уровня с уровнями {1,5,10}

```python
x = np.arange(-10, 10, 0.25)
y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(x, y)
Z = func(X,Y)
fig, ax = plt.subplots();
CS = ax.contour(X, Y, Z,[1,5,10]);
ax.clabel(CS,inline=True, fontsize=8);
```
