### Задание 1.  
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
The Moscow Times - Wednesday, October 2, 2002
The Guardian - Friday, 11.10.13
Daily News - Thursday, 18 August 1977

```python
mt = 'The Moscow Times - Wednesday, October 2, 2002'    # The Moscow Times
tg = 'The Guardian - Friday, 11.10.13'    # The Guardian
dn = 'Daily News - Thursday, 18 August 1977'    # Daily News


from datetime import datetime


# The Moscow Times

mt_l = mt.split('-')
date_mt = datetime.strptime((mt_l)[1], ' %A, %B %d, %Y')
print(date_mt)

# The Guardian

tg_l = tg.split('-')
date_tg = datetime.strptime((tg_l)[1], ' %A, %d.%m.%y')
print(date_tg)

# Daily News

dn_l = dn.split('-')
date_dn = datetime.strptime((dn_l)[1], ' %A, %d %B %Y')
print(date_dn)
```

### Задание 2.  
Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:
stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]

Напишите функцию, которая проверяет эти даты на корректность. Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата).

```python
stream = ['2018-04-02', '2018-02-29', '2018-19-02']

from datetime import datetime

def date_check(dates):
    check_list = []
    for i in dates:
        try:
            datetime.strptime(i, '%Y-%m-%d') 
            #print('True')
            check_list.append(True)
        except:
            #print('False')
            check_list.append(False)
    
    return check_list
    

    
date_check(stream)
```

### Задание 3.  
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. Даты должны вводиться в формате YYYY-MM-DD. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
start_date = '2004-01-16'
end_date = '2004-06-18'


from datetime import timedelta

def period(start, end):
    
    list_period=[]
    
    try:
        current_date = datetime.strptime(start,'%Y-%m-%d')

        while current_date <= datetime.strptime(end,'%Y-%m-%d'):
            list_period.append(current_date.strftime('%Y-%m-%d'))

            current_date += timedelta(days=1)
    except:
        
        pass
        
    return list_period


period(start_date, end_date)
```
