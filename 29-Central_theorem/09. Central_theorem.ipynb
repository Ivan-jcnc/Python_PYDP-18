### Домашнее задание №9. Центральная предельная теорема и статистика

Проведите эксперименты с центральной предельной теоремой для различных распределений.  
Создайте случайную величину из выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистограммы средних этих выборок.  
При каком значении n полученное распредление похоже на нормальное?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
size = [5, 10, 25, 70, 200, 1000]

for n in size:
    means = []
    for i in range(1000):
        re = np.random.exponential(scale=1, size=n)
        means.append(re.mean())

    plt.hist(means, bins=30)
    plt.show()
```

Начиная с n=70 и выше распределение средних становится похожим на нормальное
