### Задание 1  

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:  

если источник traffic_source равен yandex или google, то в source_type ставится organic,  
для источников paid и email из России - ставим ad,  
для источников paid и email не из России - ставим other,  
все остальные варианты берем из traffic_source без изменений.

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log_2.csv', sep = ';')
```

```python
log
```

```python
def new_row(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif (row['traffic_source'] == 'email' or row['traffic_source'] == 'paid') and row['region'] == 'Russia':
        return 'ad'
    elif (row['traffic_source'] == 'email' or row['traffic_source'] == 'paid') and row['region'] != 'Russia':
        return 'other'
    else:
        return row['traffic_source']
```

```python
log['source_type'] = log.apply(new_row, axis = 1)
```

### Задание 3  

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()[['userId', 'timestamp']] # подсчет количества элементов в группе
                                                                                        # userId
ratings_count
```

```python
ratings_100 = ratings_count[ratings_count['timestamp'] > 100] # условие вывода строк с количеством оценок > 100

ratings_100
```

```python
ratings_joined = ratings_100.merge(ratings, how = 'left', on = 'userId') # соединение с основной таблицей
                                                                        # для поиска min и max значений timestamp
```

```python
ratings_joined
```

```python
ratings_max = ratings_joined.groupby('userId').max().reset_index() # поиск максимальных значений
```

```python
ratings_max.rename(columns={'timestamp_y': 'timestamp_max'}, inplace=True)
```

```python
ratings_min = ratings_joined.groupby('userId').min().reset_index() # поиск минимальных значений
```

```python
ratings_min.rename(columns={'timestamp_y': 'timestamp_min'}, inplace=True)
```

```python
ratings_max.rename(columns={'timestamp_x': 'count'}, inplace=True)
```

```python
ratings_max['timestamp_min'] = ratings_min['timestamp_min'] # добавление столбца с минимальным значением timestamp
```

```python
ratings_max['avg'] = ratings_max['timestamp_max'] - ratings_min['timestamp_min'] # добавление столбца с искомым временем жизни
```

```python
ratings_max
```
